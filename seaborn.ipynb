# Seaborn - Statistical Tool

## 1. What is Seaborn ?

Seaborn is a Python Library that builts on the top of matplotlib that makes the statistical data more easy and beautiful.

### Why we need it ?


1. For creating plots like violin , bar , pair plots.

2. Automatically handles data from pandas dataframes.

### Where we Use it ?

1. Data Analysis and Reporting.

2. Exploring relationships between variables.

3. Machine Learning , for visualizing datasets and results

## 2. Importing  Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
#3. Load Titanic dataset
df = sns.load_dataset("titanic")
df.head()
 
```

```python
## 4. Histograms & Countplots
# Histogram
sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title("Histogram of Age")
plt.show()
 
# Countplot
sns.countplot(x='class', data=df)
plt.title("Count of Passengers by Class")
plt.show()
```

```python

### 5. Scatter Plots
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.title("Age vs Fare")
plt.show()
```

```python
### 6. Boxplot & Violin Plot
# Boxplot
sns.boxplot(x='class', y='age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
 
# Violin plot
sns.violinplot(x='class', y='age', hue='sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
```

```python
### 7. Pairplot
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

```python
### 8. Heatmap
# Correlation heatmap (numeric columns only)
corr = iris.select_dtypes(include=['float64', 'int64']).corr()
 
import seaborn as sns
import matplotlib.pyplot as plt
 
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

```python
### 9. Customization & Themes
sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['age'].dropna(), bins=10, color='orange')
plt.title("Customized Histogram with Seaborn")
plt.show()
```

```python
sns.barplot(x='class', y='fare', data=df, hue='sex', errorbar=None)
plt.title("Average Fare by Class and Sex")
plt.show()
 
```

```python
"Show trends or changes in mean values across categories."
sns.pointplot(x='class', y='fare', data=df, hue='sex', markers=['o', 'x'])
plt.title("Point Plot of Fare by Class and Sex")
plt.show()
```

```python
sns.kdeplot(df['fare'].dropna(), fill=True, color='teal')
plt.title("KDE Plot of Fare Distribution")
plt.show()
```

use the Titanic dataset
 

```python
df = sns.load_dataset("titanic")
plt.figure(figsize=(8,6))
sns.barplot(x="class", y="survived", hue="sex", data=df, errorbar=None, palette="Set2")
plt.title("Passenger Survival Rate by Class and Gender")
plt.xlabel("Passenger Class")
plt.ylabel("Survival Rate")
plt.legend(title="Gender")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(df["age"], bins=30, kde=True, color="skyblue")
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="class", y="fare", hue="class", data=df, palette="coolwarm",legend=False)
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare ($)")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="age", y="fare", data=df, hue="class", style="sex", palette="Set1")
plt.title("Relationship Between Age and Fare")
plt.xlabel("Age")
plt.ylabel("Fare ($)")
plt.legend(title="Class")
plt.show()
```

```python
plt.figure(figsize=(8,6))
corr = df.select_dtypes(include=['number']).corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Titanic Numeric Features")
plt.show()
```
